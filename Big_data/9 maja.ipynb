```
import pyspark.sql.functions as f
import numpy as np
```

```
d.groupBy("Churn").mean("MonthlyCharges")\
  .show()
```

```
#number of customers | Avg Monthly | maxMonthly | avgTotal

d.groupBy("Churn")\
  .agg(f.bround(f.mean("MonthlyCharges"), 2).alias("Avg Monthly"), 
       f.count("customerID").alias("number of customers"), 
       f.max("MonthlyCharges"),
       f.bround(f.mean("TotalCharges"), 2))\
  .show()
```

```
d.groupBy("Partner")\
  .agg(f.bround(f.mean("MonthlyCharges"), 2).alias("Avg Monthly"), 
       f.count("customerID").alias("number of customers"), 
       f.max("MonthlyCharges"),
       f.bround(f.mean("TotalCharges"), 2))\
  .show()
```

```
def customer_segment(x):
  if x is None:
    return "nd"
  elif x < 1000:
    return "low"
  elif (x>= 1000) & (x < 2500):
    return "medium"
  else:
    return "high"
```

```
customer_segment(2400)
```

```
from pyspark.sql.types import *
```

```
d = d.withColumn("TotalChargesD", d.TotalCharges.cast(DoubleType()))
```

```
udf_customer_segment = udf(customer_segment, StringType())
```

```
d = d.withColumn("Segment", udf_customer_segment(d.TotalChargesD))
```

```
d.filter(d.TotalChargesD.isNull()).count()
```

```
d.groupBy("Segment").count().show()
```
